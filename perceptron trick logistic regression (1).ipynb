```python
from sklearn.datasets import make_classification
```

```python
import numpy as np
```

```python
x,y= make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(x,y):
    
    x = np.insert(x,0,1,axis=1)   # there will be a extra column that has vvalue 1 i every row
    weights = np.ones(x.shape[1])   # this will be 3 after adding extra column
    lr=0.1
    for i in range(1000):
        j = np.random.randint(0,100)     # this will take any random point on which we will do prediction
        y_hat = step(np.dot(x[j],weights))   # weights is initially what we assume like in this it will be 1 1 1
        weights = weights + lr*(y[j]-y_hat)*x[j]
        
    return weights[0],weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
intercept_,coef_ = perceptron(x,y)   # this has given our intercept and coef value
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
#np.linspace(start, stop, num) generates num evenly spaced numbers between start and stop.
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)   #Helps to zoom or crop the plot to a particular vertical range.

```
